```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
plt.figure(figsize=(8, 5), dpi=80)
x1 = iris['PetalLength'][1:50].values
y1 = iris['PetalWidth'][1:50].values
plt.scatter(x1,y1)
x2 = iris['PetalLength'][51:100].values
y2 = iris['PetalWidth'][51:100].values
plt.scatter(x2,y2)
x3 = iris['PetalLength'][101:].values
y3 = iris['PetalWidth'][101:].values
plt.scatter(x3,y3)
plt.xlabel('PetalWidth',fontsize = 20)
plt.ylabel('PetalLength', fontsize = 20)
plt.legend(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),fontsize = 20)
plt.savefig('1.png')
```

```python
plt.figure(figsize=(14, 8), dpi=80)
x1 = iris['SepalLength'][1:50].values
y1 = iris['SepalWidth'][1:50].values
plt.scatter(x1,y1)
x2 = iris['SepalLength'][51:100].values
y2 = iris['SepalWidth'][51:100].values
plt.scatter(x2,y2)
x3 = iris['SepalLength'][101:].values
y3 = iris['SepalWidth'][101:].values
plt.scatter(x3,y3)
plt.xlabel('SepalWidth',fontsize = 20)
plt.ylabel('SepalLength', fontsize = 20)
plt.legend(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),fontsize = 20)
plt.savefig('2.png')
```
